```python
# Loading data from .csv file url

# Datasource: https://github.com/t-davidson/hate-speech-and-offensive-language

import pandas as pd
import requests
import io
import pandas as pd
import requests
import io

#url = 'https://datahub.io/world-bank/se.prm.nenr/r/data.csv'
url = 'https://raw.githubusercontent.com/t-davidson/hate-speech-and-offensive-language/master/data/labeled_data.csv'

content = requests.get(url).content
df = pd.read_csv(io.StringIO(content.decode('utf-8'))).iloc[:,1:]

print('Shape:', df.shape)
df.head(4)
```

```python
# Loading data from .xlsx table file url

# Datasource: https://ndownloader.figshare.com/files/13724144

import pandas as pd
import requests
import io


url = 'https://ndownloader.figshare.com/files/13724144'
df = pd.read_excel(url)

df = df.iloc[:,1:] # drop doubled index column.
print('Shape:', df.shape); df.head(4)
```
